```python
import requests
```

```python
response = requests.get("https://search.unbxd.io/fb853e3332f2645fac9d71dc63e09ec1/demo-unbxd700181503576558/search?&q=*")
```

```python
json_response = response.json()
```

```python
print(type(json_response))
#We have a dictionary of the API
```

```python
response1 = json_response["response"]
#Extracting product from response
products = response1["products"]
```

```python
#removing duplicates and Making true and false to Yes and No
def remove_duplicates(payload):
    
    refined = [] 
    for i in payload: 
        if i not in refined: 
            refined.append(i)
    ["NO" if x == 'false' else x for x in refined]
    ["YES" if x == 'true' else x for x in refined]
    return refined

    
            
            
```

```python
new_product = []

for each in products:
    refined_dict={}

    for every_key in each.keys():  
        if isinstance(each[every_key], list):    
            refined=remove_duplicates(each[every_key])
            refined_dict[every_key]=refined
        else:
            refined_dict[every_key]=each[every_key]
    new_product.append(refined_dict)
```

```python
print(refined_dict)
```

```python
#Converting array to string separated by ","
refined_string = ','.join(map(str, refined_dict))
```

```python
print(refined_string)
```

```python
#print(refined_dict['unbxd_color_for_category'])
def colour(refined_dict):
    for items in refined_dict['unbxd_color_for_category']:
        refined_dict['unbxd_color_for_category'] = items.split("::")[0]
```

```python
for refined_dicts in new_product:
    colour(refined_dicts)
```
